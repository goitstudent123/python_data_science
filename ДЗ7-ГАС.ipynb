У партії з 50 деталей 5 нестандартних. Визнач ймовірність того, що серед обраних навмання для перевірки шести деталей дві виявляться нестандартними. Розв’яжи аналітично.

Нехай
(N=50),
(K=5) - число нестандартних,
(n=6) - розмір вибірки.

тоді, імовірність рівно двох нестандартних:

$$
\mathbb{P}\{X=2\}
= \frac{\binom{K}{2}\binom{N-K}{\,n-2\,}}{\binom{N}{n}}
= \frac{\binom{5}{2}\binom{45}{4}}{\binom{50}{6}}.
$$

На площині накреслені дві концентричні окружності (мають спільний центр), радіуси яких 5 і 10 см відповідно. Знайди ймовірність того, що точка, кинута навмання у велике коло, потрапить також у кільце, утворене побудованими окружностями.**bold text**

Нехай радіуси \(r_1=5\), \(r_2=10\). Шукана імовірність — відношення площ:
$$
\mathbb{P}
= \frac{\pi r_2^2 - \pi r_1^2}{\pi r_2^2}
= \frac{10^2 - 5^2}{10^2}
= \frac{100-25}{100}=\frac{75}{100}=0.75.
$$

**Завдання 6**

Урна містить одну кулю, про яку відомо, що вона або біла, або чорна з однаковими ймовірностями. В урну кладуть білу кулю і потім навмання виймають одну кулю. Вона виявилася білою. Яка ймовірність того, що куля, яка залишилася, є білою?

Позначимо \(W\) — «початкова куля біла», \(B\) — «початкова куля чорна», \(E\) — «витягнули білу».

Маємо $$ \mathbb{P}(W)=\mathbb{P}(B)=\tfrac12, \mathbb{P}(E\mid W)=1, \mathbb{P}(E\mid B)=\tfrac12
$$
Тоді
$$
\mathbb{P}(W\mid E)
=\frac{\mathbb{P}(E\mid W)\mathbb{P}(W)}{\mathbb{P}(E\mid W)\mathbb{P}(W)+\mathbb{P}(E\mid B)\mathbb{P}(B)}
=\frac{\tfrac12}{\tfrac12+\tfrac14}
=\frac{2}{3}.
$$

Якщо початкова куля була біла (W), у урні після витягання білої залишається біла. Отже,
$$
\mathbb{P}(\text{залишилась біла}\mid E)=\mathbb{P}(W\mid E)=\frac{2}{3}.
$$

**Завдання 7

Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1.
У початковий момент часу t=0 ціна рівна 0.
Визнач середню ціну акції через 3 одиниці часу (t=3).

За крок \(X_i\) приймає значення \(+2\) або \(-1\) з імовірностями \(1/2\) та \(1/2\).
$$
\mathbb{E}[X_i]=\tfrac12\cdot 2+\tfrac12\cdot(-1)=\tfrac12.
$$
Через 3 кроки \(S_3=X_1+X_2+X_3\), тому
$$
\mathbb{E}[S_3]=3\cdot \mathbb{E}[X_i]=3\cdot\tfrac12=1.5.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate(n, steps=3, seed=0):
    rng = np.random.default_rng(seed)
    # +2 with p=0.5, -1 with p=0.5, for each of `steps`
    steps_mat = rng.choice([2, -1], size=(n, steps))
    s = steps_mat.sum(axis=1)
    return s

for n in [10, 100, 1000, 10000]:
    s = simulate(n, steps=3, seed=42)
    print(f"n={n}  mean={s.mean():.4f}")
    plt.figure(figsize=(6,3.2))
    plt.hist(s, bins=np.arange(s.min()-0.5, s.max()+1.5, 1), edgecolor='black')
    plt.title(f"t=3, histogram, n={n}")
    plt.xlabel("price")
    plt.ylabel("count")
    plt.tight_layout()
    plt.show()
```
