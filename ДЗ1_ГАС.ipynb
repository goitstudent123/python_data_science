# Завдання 1.
В шаблоні наведені числові вектори $\overline{a}$ та $\overline{b}$. Порахуй наступне:
- сума $\overline{a}$ та $\overline{b}$
- різниця $\overline{a}$ та $\overline{b}$
- сума $\overline{a}$ та $\overline{b}^T$. Поясни отриманий результат.
- матричний добуток (dot product) $\overline{a}$ та $\overline{b}^T$.
- матричний добуток (dot product) $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- добуток Адамара (Hadamard product) $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- ділення $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- ділення $\overline{a}$ та $\overline{b}^T$. Поясни отриманий результат.

````python
import numpy as np
# Код оголошення векторів не можна змінювати
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])
````

````python
res = a + b
res
````

````python
res = a - b
res
````

````python
# Пояснення: вектор (1х5) додається до вектору (5х1) отримуємо матрицю 5x5,
#            де кожен рядок це вектор a, і до кожного з значень кожного рядка
#            додаєтся відповідний елемент вектора b, для першого 0.5, для другого 1 і тд
res = a + b.T
res
````

````python
res = a @ b.T
res
````

````python
# Пояснення: помилка через розмірність (1х5) @ (1x5),
# очікується що кількість рядків другого вектору дорівнює кількості колонок першого
res = a @ b
res
````

````python
# Пояснення: [[1*0.5, 2*1, 3*2, 4*3, 5*4]]
res = a * b
res
````

````python
# Пояснення: [[1/0.5, 2/1, 3/2, 4/3, 5/4]]
res = a / b
res
````

````python
# Пояснення: вектор (1х5) ділиться на вектору (5х1) отримуємо матрицю 5x5,                      [1/0.5, 2/0.5 ...]
#            де кожен рядок це вектор a, кожний з улуменів кожного рядка                        [1/1,   2/1   ...]
#            ділиться на відповідний елемент вектора b, для першого 0.5, для другого 1 і тд     [1/2,   2/2   ...]
res = a / b.T
res
````

# Завдання 2 (Альтернативне)

````python
import cv2 as cv
import urllib
import numpy as np
from google.colab.patches import cv2_imshow as cv_imshow

# Read an image
def read_image_by_url(url):
    req = urllib.request.urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv.imdecode(arr, -1)
    return img

url = 'https://s3.dualstack.us-east-2.amazonaws.com/pythondotorg-assets/media/community/logos/python-logo-only.png'

img = read_image_by_url(url)
cv_imshow(img)
````

````python
# prepare:
image = img.copy()
h, w = image.shape[:2]

# 1. scale
current = image.copy()

def center_transform(matrix):
    center = np.array([[1, 0, -w/2], [0, 1, -h/2], [0, 0, 1]])
    uncenter = np.array([[1, 0, w/2], [0, 1, h/2], [0, 0, 1]])
    return uncenter @ matrix @ center

def scaling_matrix(sx, sy):
    return np.array([
        [sx, 0,  0],
        [0,  sy, 0],
        [0,  0,  1]
    ])

M1 = center_transform(scaling_matrix(0.5, 3))
current = cv.warpPerspective(current, M1, (w, h), borderMode=cv.BORDER_REPLICATE)
cv_imshow(current)
````

````python
# 2. reflect
current = image.copy()
def reflection_origin():
    return np.array([
        [-1,  0, 0],
        [0,  -1, 0],
        [0,   0, 1]
    ])
M2 = center_transform(reflection_origin())
current = cv.warpPerspective(current, M2, (w, h), borderMode=cv.BORDER_REPLICATE)
cv_imshow(current)
````

````python
# 3. translate
current = image.copy()
def translation_matrix(tx, ty):
    return np.array([
        [1, 0, tx],
        [0, 1, ty],
        [0, 0, 1]
    ])


M3 = center_transform(translation_matrix(-3, 1))
current = cv.warpPerspective(current, M3, (w, h), borderMode=cv.BORDER_REPLICATE)
cv_imshow(current)
````

````python
# 4. shear
current = image.copy()
def shear_oy(angle_deg):
    t = np.tan(np.radians(angle_deg))
    return np.array([
        [1, t, 0],
        [0, 1, 0],
        [0, 0, 1]
    ])

M4 = center_transform(shear_oy(60))
current = cv.warpPerspective(current, M4, (w, h), borderMode=cv.BORDER_REPLICATE)
cv_imshow(current)
````

````python
# 5. rotation:
current = image.copy()
def rotation_matrix(angle_deg):
    theta = np.radians(angle_deg)
    return np.array([
        [ np.cos(theta), -np.sin(theta), 0],
        [ np.sin(theta),  np.cos(theta), 0],
        [0, 0, 1]
    ])

M4 = center_transform(rotation_matrix(30))
current = cv.warpPerspective(current, M4, (w, h), borderMode=cv.BORDER_REPLICATE)
cv_imshow(current)
````

````python

# 6. 1 (scale) → 2 (reflect) → 4 (shear) → 5 (rotate)
M = rotation_matrix(30) @ shear_oy(60) @ reflection_origin() @ scaling_matrix(0.5, 3)

M_centered = center_transform(M)

output = cv.warpPerspective(image, M_centered, (w, h), borderMode=cv.BORDER_REPLICATE)

cv_imshow(output)
````
